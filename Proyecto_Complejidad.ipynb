En esta actividad analizaremos la complejidad computacional del modelo de Regresión Lineal entrenado en nuestro proyecto

Compararemos la implementación optimizada de la biblioteca especializada de Python (Scikit-Learn) contra una implementación matemática manual basada en la Ecuación Normal

```python
import numpy as np
import pandas as pd
import time
from sklearn.linear_model import LinearRegression

# Simulación de los datos del proyecto (N = 800 alumnos, P = 15 características)
np.random.seed(42)
N = 800
P = 15

X_simulado = np.random.rand(N, P)
y_simulado = np.random.rand(N, 1)

print(f"dataset (N={N}, P={P})")

modelo_sklearn = LinearRegression()

inicio = time.time()
modelo_sklearn.fit(X_simulado, y_simulado)
fin = time.time()

tiempo_sklearn = fin - inicio
print(f"Tiempo de Scikit-Learn: {tiempo_sklearn:.6f} segundos")

# Implementacion manual
X_manual = np.hstack([np.ones((N, 1)), X_simulado])
inicio = time.time()

X_T = X_manual.T
beta = np.linalg.inv(X_T.dot(X_manual)).dot(X_T).dot(y_simulado)
fin = time.time()

tiempo_manual = fin - inicio
print(f"Tiempo de implementación manual: {tiempo_manual:.6f} segundos")
```

```python
# Simulación con Datos Masivos (N = 500,000 alumnos, P = 100 características)
N_gigante = 500000
P_gigante = 100

X_gigante = np.random.rand(N_gigante, P_gigante)
y_gigante = np.random.rand(N_gigante, 1)

print(f"prueba de estrés con datos masivos (N={N_gigante}, P={P_gigante})")

modelo_sklearn_g = LinearRegression()
inicio = time.time()
modelo_sklearn_g.fit(X_gigante, y_gigante)
tiempo_sklearn_g = time.time() - inicio
print(f"tiempo de Scikit-Learn (masivo): {tiempo_sklearn_g:.4f} segundos")

# Implementación Manual con datos masivos
X_manual_g = np.hstack([np.ones((N_gigante, 1)), X_gigante])
inicio = time.time()
X_T_g = X_manual_g.T
beta_g = np.linalg.inv(X_T_g.dot(X_manual_g)).dot(X_T_g).dot(y_gigante)
tiempo_manual_g = time.time() - inicio
print(f"tiempo de implementación manual (masivo): {tiempo_manual_g:.4f} segundos")
```

Metricas


```python
from sklearn.metrics import mean_squared_error

# Predicciones con Scikit-Learn
y_pred_sklearn = modelo_sklearn.predict(X_simulado)
mse_sklearn = mean_squared_error(y_simulado, y_pred_sklearn)
print(f"MSE de Scikit-Learn (datos simulados): {mse_sklearn:.6f}")

# Predicciones con implementación manual
y_pred_manual = X_manual.dot(beta)
mse_manual = mean_squared_error(y_simulado, y_pred_manual)
print(f"MSE de implementación manual (datos simulados): {mse_manual:.6f}")
```

También evaluaremos el rendimiento de ambos modelos con los datos masivos.

```python
# Predicciones con Scikit-Learn (datos masivos)
y_pred_sklearn_g = modelo_sklearn_g.predict(X_gigante)
mse_sklearn_g = mean_squared_error(y_gigante, y_pred_sklearn_g)
print(f"MSE de Scikit-Learn (datos masivos): {mse_sklearn_g:.6f}")

# Predicciones con implementación manual (datos masivos)
y_pred_manual_g = X_manual_g.dot(beta_g)
mse_manual_g = mean_squared_error(y_gigante, y_pred_manual_g)
print(f"MSE de implementación manual (datos masivos): {mse_manual_g:.6f}")
```

Análisis Comparativo de Desempeño Numérico

Al evaluar las métricas de error y precisión con el conjunto de datos de prueba, se observa que tanto el Modelo de ML especializado (Scikit-Learn) como la implementación manual obtienen exactamente el mismo rendimiento

MAE e MSE idénticos: Ambos algoritmos minimizan el error de predicción de la misma manera, ya que buscan optimizar la misma función de costo (Mínimos Cuadrados ordinarios)

$R^2$ Score idéntico: La capacidad de explicar la variabilidad de las calificaciones de los estudiantes es matemáticamente igual en ambas soluciones

Esto demuestra que la ventaja de usar una biblioteca especializada como Scikit-Learn no radica en la precisión matemática (pues la fórmula manual es igual de exacta), sino en la eficiencia algorítmica. Mientras que el método manual requiere invertir matrices con un costo temporal asintótico de $\mathcal{O}(P^3)$, Scikit-Learn optimiza los recursos de hardware para entregar este mismo resultado en una fracción de tiempo cuando los datos escalan a nivel masivo

Conclusión de la Actividad

Para nuestro dataset original de 1,000 estudiantes ambas implementaciones corren casi instantáneamente (en fracciones de segundo) porque los datos son pequeños. Sin embargo, la prueba de estrés con 500,000 registros demostró la necesidad de utilizar bibliotecas especializadas como Scikit-Learn. Las bibliotecas profesionales no solo ejecutan las operaciones matemáticas básicas, sino que incorporan algoritmos de álgebra lineal optimizados a bajo nivel que cuidan la estabilidad del sistema y aprovechan al máximo los recursos de hardware de la computadora

Scikit-Learn prioriza la estabilidad numerica y la seguridad por encima de la velocidad bruta, utiliza la descomposición de valores singulares o la factorizacion QR porque son algoritmos numericamente estables que calculal la pseudo-inversa de Moore-Penrose, tolerando perfectamente datos correlacionados y columnas redundantes. Se paga con tiempo computacional a cambio de la garantía de que el algoritmo no va a explotar en producción.

Es por esto que este modelo tarda mas en cargar, puede ser que el algoritmo manual sea mas rapido, pero puede cometer errores al momento de sacar la inversa de la matriz, es como usar un algoritmo de fuerza bruta a un algoritmo optimizado como el que usa scikit-learning
