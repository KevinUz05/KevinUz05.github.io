**Análisis matemático del modelo**

A. Formulación Matemática del Modelo

El modelo de Regresión Lineal Múltiple asume que la variable dependiente Y puede expresarse como una combinación lineal de un conjunto de q variables independientes (regresores) más un componente de error aleatorio.
De forma explícita para cada observación i (donde i=1,2,…,n):
$Y_i = \beta_0 + \beta_1 X_{i1} + \beta_2 X_{i2} + \cdots + \beta_q X_{iq} + \epsilon_i$

Para tratar el modelo computacional y analíticamente, se recurre a la notación matricial:
$Y = X\beta + \epsilon$

Donde las dimensiones de las matrices constituyentes son las siguientes:
*   $Y$ es un vector de respuestas de dimensiones $n \times 1$.
*   $X$ es la matriz de diseño de dimensiones $n \times (q+1)$, donde la primera columna es un vector de unos (1) que permite modelar el intercepto $\beta_0$.
*   $\beta$ es el vector de coeficientes de dimensiones $(q+1) \times 1$, representando los parámetros a estimar.
*   $\epsilon$ es el vector de perturbaciones o residuos aleatorios de dimensiones $n \times 1$.

Supuestos fundamentales de la inferencia clásica:
*   **Linealidad:** La relación entre las expectativas de $Y$ y los regresores es lineal en los parámetros $\beta$.
*   **Homocedasticidad:** La varianza de los errores es constante para cualquier nivel de las variables explicativas: $Var(\epsilon_i) = \sigma^2 \ \forall i$.
*   **Independencia (No autocorrelación):** Los errores de observaciones distintas no están correlacionados entre sí: $Cov(\epsilon_i, \epsilon_j) = 0 \ \forall i \neq j$.
*   **Normalidad:** Los errores siguen una distribución gaussiana: $\epsilon \sim N(0, \sigma^2 I_n)$, lo cual convalida las pruebas de hipótesis sobre los coeficientes.
*   **Ausencia de multicolinealidad perfecta:** $Rango(\mathbf{X}) = q+1 \le n$, lo que implica que ninguna variable independiente es una combinación lineal perfecta de las otras.

B. Función de Costo

La optimización del modelo busca minimizar las desviaciones verticales entre los puntos observados y el hiperplano de regresión ajustado. La métrica por excelencia para este propósito es la suma de los residuos al cuadrado, la cual formaliza el Error Cuadrático Medio (MSE) como función de pérdida:
$J(\beta) = MSE(\beta) = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2 = \frac{1}{n} (Y - X\beta)^T (Y - X\beta)$

Expandiendo el producto matricial de la función de costo:
$J(\beta) = \frac{1}{n} (Y^T Y - 2\beta^T X^T Y + \beta^T X^T X \beta)$

C. Método de Optimización (Mínimos Cuadrados Ordinarios - OLS)

Para deducir el estimador óptimo de $\beta$ que minimice la pérdida $J(\beta)$, se computa el gradiente con respecto al vector de parámetros e iguala a cero (0):
$\frac{\partial J(\beta)}{\partial \beta} = \frac{1}{n} (-2X^T Y + 2X^T X \beta) = 0$

Al resolver este sistema analítico se derivan las célebres Ecuaciones Normales:
$X^T X \beta = X^T Y$

Si la matriz $X^T X$ posee rango completo, es invertible, proporcionando una solución única en forma cerrada para el vector de coeficientes estimados ($\hat{\beta}$):
$\hat{\beta} = (X^T X)^{-1} X^T Y$

A nivel analítico, este estimador funciona porque proyecta de forma ortogonal el vector de respuestas $Y$ sobre el espacio de columnas generado por la matriz de diseño $X$. De acuerdo con el Teorema de Gauss-Márkov, dentro de la clase de estimadores lineales e insesgados, el estimador OLS es el de mínima varianza (BLUE: Best Linear Unbiased Estimator).

D. Propiedades Matemáticas del Modelo

*   **Convexidad:** La función de costo $J(\beta)$ es estrictamente convexa si $X^T X$ es definida positiva. Esto garantiza matemáticamente que cualquier mínimo local es también un mínimo global, impidiendo que los algoritmos queden atrapados en soluciones subóptimas.
*   **Solución Cerrada:** A diferencia de modelos complejos (como redes neuronales artificiales), la regresión lineal no requiere obligatoriamente de optimización iterativa estocástica para converger; se resuelve directamente mediante álgebra lineal si las dimensiones lo permiten.
*   **Naturaleza Probabilística vs. Determinista:** El modelo no es puramente matemático/determinista (como $y=mx+b$ sin perturbación). Al incorporar el término $\epsilon_i$, se asume un componente estocástico no controlado (ruido gaussiano), lo que transforma el mapeo en una distribución condicional: $Y_i | X_i \sim N(X_i \beta, \sigma^2)$.
*   **Dimensionalidad:** El comportamiento del estimador depende críticamente de la relación entre el número de observaciones (n) y el número de variables (q). Si $q>n$, el sistema está subdeterminado, $X^T X$ se vuelve singular (no invertible) y la regresión lineal OLS clásica no puede calcularse sin regularización (Ridge/Lasso).

E. Pros y Contras desde una Perspectiva Matemática

**Ventajas:**
*   **Ausencia de mínimos locales anómalos:** La convexidad de la función MSE asegura la convergencia al óptimo global de manera unívoca.
*   **Solución cerrada directa:** No sufre de problemas de sintonización de hiperparámetros de aprendizaje (learning rate) si se computa de forma matricial directa.
*   **Alta interpretabilidad:** Cada coeficiente $\beta_j$ representa el cambio esperado en $Y$ por cada unidad de incremento en $X_j$, manteniendo constantes los demás regresores.

**Desventajas:**
*   **Sensibilidad extrema a Outliers:** Al cuadrar los residuos en la función de costo, las observaciones atípicas ejercen un "apalancamiento" desproporcionado sobre la derivada del gradiente, distorsionando el hiperplano óptimo.
*   **Vulnerabilidad a la Multicolinealidad:** Si dos o más variables independientes están altamente correlacionadas, el determinante de $X^T X$ se aproxima a cero. Esto infla drásticamente los elementos de la diagonal de la matriz de varianza-covarianza $(X^T X)^{-1} \sigma^2$, haciendo que las estimaciones de los parámetros sean sumamente inestables.
*   **Costo Computacional de la Inversión:** Calcular $(X^T X)^{-1}$ mediante algoritmos estándar requiere un tiempo computacional de orden $O(q^3)$, lo cual es ineficiente si el número de variables $q$ escala a decenas de miles.

F. Argumento Teórico de Adecuación

La regresión lineal múltiple es el enfoque óptimo inicial para el dataset estudiado debido a la naturaleza continua de la variable de respuesta CGPA. Dado que disponemos de un tamaño muestral significativamente mayor al número de regresores (n≫q), el riesgo de caer en matrices singulares disminuye drásticamente.

**Aplicación Práctica Modelo**

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import joblib

# 1. Carga y división de estructuras matriciales
df = pd.read_csv("student_lifestyle_performance_dataset.csv")
X = df.drop(columns=["CGPA"])
y = df["CGPA"]

# 2. Configuración del preprocesamiento matricial
categorical_features = ["Branch", "Diet_Type", "Residence"]

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ],
    remainder="passthrough" # Mantiene intactas las variables numéricas: Internal_Marks, Study_Hours, etc.
)

# 3. Construcción del Pipeline de Machine Learning
pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LinearRegression())
])

# 4. Ajuste y entrenamiento del modelo completo
pipeline.fit(X, y)

# 5. Exportación del artefacto entrenado
joblib.dump(pipeline, "model_pipeline.pkl")
print("¡Pipeline exportado correctamente!")
```

**Discusión**

Herramientas: scikit-learn se orienta a la predicción y el despliegue eficiente (producción), mientras que statsmodels se enfoca en el diagnóstico riguroso de supuestos estadísticos (análisis de residuos y control de multicolinealidad mediante el VIF).
En el caso específico del rendimiento estudiantil, variables altamente correlacionadas entre sí (por ejemplo, el porcentaje de asistencia Attendance_Percentage y las notas parciales Internal_Marks) exigen un monitoreo constante del Factor de Inflación de la Varianza (VIF) para evitar que los coeficientes individuales pierdan estabilidad analítica o presenten signos contraintuitivos debido a la multicolinealidad.
